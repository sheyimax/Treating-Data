## Filling the Null Values Method

```python
housing_1 = housing.fillna(value=0)      #Using the fillna() syntax

housing_1.head()
```

```python
#Lets check if there is still more missing value in the data

housing_1.isnull().sum().sum()
```

```python
# Filling Null values with previous values

housing_2 = housing.fillna(method = 'pad')

housing_2.head(10)
```

```python
#To check if all null values are filled

housing_2.isnull().sum()
```

```python
# Filling the null values with the next value

housing_3 = housing.fillna(method = 'bfill')

housing_3.head(6)
```

```python
#Filling the null values column wise

housing_4 = housing.fillna(method='bfill', axis=1)  

housing_4.head(10)
```

```python
#Filling Different Values in Null in different columns

housing_5 = housing.fillna({'society': 'Ayobo',
                           'balcony': 98})

housing_5
```

```python
#Filling the null values with the mean of the column

housing_6 = housing.fillna(value=housing['balcony'].mean())

housing_6
```

```python
#Filling the null value with the max of the column

housing_7 = housing.fillna(value=housing['balcony'].max())

housing_7
```

## Droping such rows using dropna()

```python
#lets print the housing data again

housing
```

```python
#using the dropna method

housing_8 = housing.dropna()

housing_8
```

```python
#using the args in dropna(); any or all

housing_9 = housing.dropna(how='any')  #This drop any rows with a null value
                                       #all drop rows with full missing value
housing_9
```

## Using the Replacing Method

```python
housing_10 = housing.replace(to_replace=np.nan, value=123456)   #np.nan is a numpy syntax to represent all the null values

housing_10
```

```python
#We want to replace the 'balcony' columns with the values 3 to 5

housing_11 = housing.replace(to_replace=3.0, value=5.0)

housing_11
```

## Using the Interpolate method

```python
#Lets interpolate the 'balcony' column

housing_12 = housing
housing_12['balcony'] = housing_12['balcony'].interpolate(method='linear')

housing_12
```
